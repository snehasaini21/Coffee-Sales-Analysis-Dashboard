RUNNING SQL QUERIES FOR DATA ANALYSIS

```python
# Total number of Transactions
query1 = '''
select count(payment_mode) as transactions from table_2024 
'''
pd.read_sql_query(query1, conn)
```

```python
# Total revenue in 2024
query3 = '''
select sum(money) as total_revenue
from table_2024
'''
pd.read_sql_query(query3, conn)
```

```python
# Unique customers (excluding cash)
query5 = '''
select count(distinct card_number) as unique_card_users_count
from table_2024 where card_number != 'cash_user' 
 '''
pd.read_sql_query(query5, conn)
```

```python
# Sales Patterns
# Sales by Month in 2024
query6 = '''
select month, 
sum(money) as Total_Revenue
from table_2024
group by(month)
order by Total_Revenue desc;
'''
pd.read_sql_query(query6, conn)
```

```python
# Sales by Weekday
query7 = '''
select day_name,
sum(money) as Revenue
from table_2024
group by(day_name)
order by Revenue desc;
'''
pd.read_sql_query(query7, conn)
```

```python
# Sales by am_pm
query8 = '''
select am_pm,
count(money) as Revenue
from table_2024
group by(am_pm)
order by Revenue desc;
'''
pd.read_sql_query(query8, conn)
```

```python
# Sales by hour
query9 = '''
select hour,
sum(money) as Revenue
from table_2024
group by(hour)
order by (Revenue) desc
'''
pd.read_sql_query(query9, conn)
```

```python
# Product Analysis
# Top 5 coffee items
query10 = '''
select coffee_name, 
count(coffee_name) Total_Count
from table_2024
group by coffee_name
order by Total_Count desc
limit 5
'''
pd.read_sql_query(query10, conn)
```

```python
# Revenue by Product
query11 = '''
select coffee_name,
sum(money) as Total_Revenue_by_Product
from table_2024
group by coffee_name
order by Total_Revenue_by_Product desc
'''
pd.read_sql_query(query11, conn)
```

```python
table_2024.columns
```

```python
# Payment Insights
# Transaction count by mode
query12 = '''
select payment_mode,
count(payment_mode) as Total
from table_2024
group by payment_mode
order by Total desc;
'''
pd.read_sql_query(query12, conn)
```

```python
# Revenue by payment mode
query13 = '''
select payment_mode,
sum(money) as Money_Earned
from table_2024
group by payment_mode
order by Money_Earned desc
'''
pd.read_sql_query(query13, conn)
```

```python
# Customer Behavior
# Top 5 customers by transactions made by card
query14 = '''
select card_number,
count(card_number) as Total_Transaction
from table_2024
where card_number != 'cash_user'
group by card_number
order by Total_Transaction desc
limit 5
'''
pd.read_sql_query(query14, conn)
```

```python
# Top 5 customers by spend
query15 = '''
select card_number,
sum(money) as Total_Spent
from table_2024
where card_number != 'cash_user'
group by card_number
order by Total_Spent desc
limit 5
'''
pd.read_sql_query(query15, conn)
```

```python
table_2024.columns
```

```python
# Time_Based Comparative Analysis
# Sales Trend by Date
query16 = '''
select date, sum(money) as daily_sales
from table_2024
group by date
order by daily_sales desc
'''
pd.read_sql_query(query16, conn)
```

```python
# Hourly Trend by Weekday
query17 = '''
select day_name, 
sum(money) as Total_Sales
from table_2024
group by day_name
order by Total_Sales desc
'''
pd.read_sql_query(query17, conn)
```

```python
# Performance Metrics
# Average Order Value
query18 = '''
select avg(money) as Average_money
from table_2024
'''
pd.read_sql_query(query18, conn)
```

```python
# Loyal Customers
query19 = '''
select card_number as loyal_customers,
count(card_number) as Customers_count
from table_2024
where card_number != 'cash_user'
group by loyal_customers
having 'Customers_count'>1
order by Customers_count desc
'''
pd.read_sql_query(query19, conn)
```

```python
# Find Customers Who Purchased Every Month
query20 = '''
select card_number
from table_2024
where card_number != 'cash_user'
group by card_number
having count(distinct month) = 12
'''
pd.read_sql_query(query20, conn)
# no customer have purchased every month
```

```python
# Get Products That Were Only Bought in the Morning(6 AM to 12 PM)
query21 = '''
SELECT coffee_name, am_pm,
count(am_pm)
FROM table_2024
WHERE hour BETWEEN 6 AND 11
group by coffee_name, am_pm
;
'''
pd.read_sql_query(query21, conn)
```

```python
# Get Products That Were Only Bought in the Evening(18 PM to 22 PM)
query22 = '''
SELECT coffee_name, am_pm,
count(am_pm)
from table_2024
where hour between 12 and 23
group by coffee_name, am_pm
'''
pd.read_sql_query(query22, conn)
```

```python
# Weekday vs Weekend Sales Volume
query23 = '''
select 
case
when day_name in ('Saturday', 'Sunady') then 'Weekend'
else 'Weekday'
end as 'day_type',
count(*) as total_transactions,
sum(money) as total_revenue
from table_2024
group by day_type
'''
pd.read_sql_query(query23, conn)
```

```python
# Which coffee has the highest average selling price?
query24 = '''
select coffee_name, round(avg(money), 2) as avg_price
from table_2024
group by coffee_name
order by avg_price desc
limit 5
'''
pd.read_sql_query(query24, conn)
```

```python
# Which hour has the highest average revenue per transaction
query25 = '''
select hour, 
count(*) as transactions,
sum(money) as total_revenue,
round(avg(money), 2) as avg_revenue_per_transaction
from table_2024
group by hour
order by avg_revenue_per_transaction desc
'''
pd.read_sql_query(query25, conn)
```
